```python
import argostranslate.translate

languages = argostranslate.translate.get_installed_languages()

english = next(lang for lang in languages if lang.code == "en")
spanish = next(lang for lang in languages if lang.code == "es")

translator = english.get_translation(spanish)

print("Traductor listo.")
```

```python
def translate(text):

    if text is None:
        return ""

    text = str(text)

    if not text.strip():
        return ""

    return translator.translate(text)
```

```python
import pandas as pd

INPUT   = "enriched_recipes.csv"
OUTPUT  = "enriched_recipes_es.csv"

df = pd.read_csv(INPUT)

print(f"Recetas encontradas: {len(df)}")

translated = []

for index, row in df.iterrows():

    if index % 100 == 0:
        print(f"{index}/{len(df)}")

    new_row = row.copy()

    new_row["recipe_name"] = translate(row["recipe_name"])
    new_row["main_ingredients"] = translate(row["main_ingredients"])
    new_row["meal_type"] = translate(row["meal_type"])
    new_row["taste_profile"] = translate(row["taste_profile"])
    new_row["served_temperature"] = translate(row["served_temperature"])
    new_row["season"] = translate(row["season"])
    new_row["difficulty"] = translate(row["difficulty"])
    new_row["characteristics"] = translate(row["characteristics"])
    new_row["semantic_summary"] = translate(row["semantic_summary"])

    new_row["recipe_id"] = row["recipe_id"]

    translated.append(new_row)

translated_df = pd.DataFrame(translated)

translated_df.to_csv(
    OUTPUT,
    index=False,
)

print("Traducción finalizada.")
```
